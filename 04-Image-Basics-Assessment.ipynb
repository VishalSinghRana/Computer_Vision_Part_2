#### TASK: Open the *dog_backpack.jpg* image from the DATA folder and display it in the notebook. Make sure to correct for the RGB order.

```python
import cv2 
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img = cv2.imread("../DATA/dog_backpack.jpg")
plt.imshow(img)
```

```python
fix_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB) # Converting the image to RGB
```

#### TASK: Flip the image upside down and display it in the notebook.

```python
new_img = cv2.flip(fix_img,0)
plt.imshow(new_img)
```

#### TASK: Draw an empty RED rectangle around the dogs face and display the image in the notebook.

```python
cv2.rectangle(fix_img,(150,350),(600,700),(255,255,0),thickness=5)
plt.imshow(fix_img)
```

```python
vertices = np.array([[400,400],[250,650],[650,650]])
pts = vertices.reshape((-1,1,2))
pts.shape
```

#### TASK: Draw a BLUE TRIANGLE in the middle of the image. The size and angle is up to you, but it should be a triangle (three sides) in any orientation.

```python
cv2.polylines(fix_img,[pts], isClosed=True, color =(255, 0 ,0),thickness=5 )
plt.imshow(fix_img)
```

```python
cv2.fillPoly(fix_img,[pts], color =(255, 0 ,0)) # fillPoly fills up the polygon figure.
plt.imshow(fix_img)
```
